## huggingface integration

```python
from langchain_community.llms import HuggingFaceHub, HuggingFaceTextGenInference, HuggingFacePipeline
from langchain_community.embeddings import HuggingFaceEmbeddings

from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate

from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# from langchain.llms import OpenAI  # for openai
```

```python
# implementing huggingface model
model_id = "MBZUAI/LaMini-Flan-T5-783M"  # EleutherAI/gpt-neo-2.7B  # MBZUAI/LaMini-Cerebras-256M  # "google/flan-t5-xxl"
tokenizer = AutoTokenizer.from_pretrained(model_id, cache_dir = "/content/gdrive/My Drive/gen_ai")
#model = AutoModelForCausalLM.from_pretrained(model_id, cache_dir = "/content/gdrive/My Drive/gen_ai")

pipe = pipeline(
    'text2text-generation', model = model_id, tokenizer = tokenizer, max_new_tokens = 512,  # device_map = 'auto'  # model_kwargs={"temperature": 0.5, "max_length": 64}
)

my_model = HuggingFacePipeline(pipeline = pipe)
```

```python
# creating prompt template using langchain
prompt_template = """rewrite the following sentence in a formal legal tone: {fact}"""

my_prompt_template = PromptTemplate(
    input_variables=["fact"],
    template= prompt_template,
)
```

```python
# creating chain using langchain
mymodel_chain = LLMChain(
    llm= my_model,
    prompt = my_prompt_template,
    output_key= "output_mymodel",
    verbose= False
)
```

```python
# calling the model/ invoking the chain
output = mymodel_chain.invoke(
    {"fact": fact}
)
print(output)
```

```python
fact = "he was accused of taking bribe."
my_prompt = my_prompt_template.format(
    fact = fact
)
```

```python
output2 = mymodel_chain.invoke(my_prompt)
print(output2)
```

```python
facts= [
    "The accused did not create or submit forged documents to the authority.",
    "Physical or mental torture.",
    "Loan pending with the bank.",
    "The defendant was seen leaving the crime scene at the time of the incident."
]

for fact in facts:
    output = mymodel_chain.invoke({"fact": fact})
    print(output)
```

## chromadb + langchain


```python
# basic
# import
import json
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.embeddings.sentence_transformer import (
    SentenceTransformerEmbeddings,
)
from langchain_community.vectorstores import Chroma

# load the document and split it into chunks
loader = PyPDFLoader("/content/gdrive/My Drive/gen_ai/two-friends-and-the-bear-story-pdf.pdf")
documents = loader.load()

# split it into chunks
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(documents)

# create the open-source embedding function
embedding_function = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

# load it into Chroma
db = Chroma.from_documents(docs, embedding_function)
print(db.get()["documents"][0])
print('>>>>>>>>>>> \n')
print(db.get()["documents"][1])


# query it
query = "moral of the story"
docs = db.similarity_search(query)

# print results
print(docs[0].page_content)
```

```python
# save to disk
db2 = Chroma.from_documents(docs, embedding_function, persist_directory="/content/gdrive/My Drive/gen_ai")
docs = db2.similarity_search(query)

# load from disk
db3 = Chroma(persist_directory="/content/gdrive/My Drive/gen_ai", embedding_function=embedding_function)
docs = db3.similarity_search(query)




print(docs[0].page_content)
```
